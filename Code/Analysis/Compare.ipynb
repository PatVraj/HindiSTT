```python
def read_file(filename):
    """Reads a file and returns its content as a list of sentences."""
    with open(filename, "r") as file:
        content = file.read()
        # Split content into sentences using periods as delimiters
        sentences = content.split(".")
        # Remove any empty sentences
        sentences = [s.strip() for s in sentences if s.strip()]
        return sentences

def compare_files(file1, file2):
    """Compares two files line by line (sentence by sentence)."""
    sentences1 = read_file(file1)
    sentences2 = read_file(file2)

    # Compare sentences
    for i, (s1, s2) in enumerate(zip(sentences1, sentences2), start=1):
        if s1 != s2:
            print(f"Difference in line {i}:")
            print(f"Master: {s1}")
            print(f"File 2: {s2}")
            print()

# Usage
master = '../Text/Translations/(Master) WER.txt'
ASR_Text = '../Text/Translations/NeuraltoAmazon.txt'
# ../Text/Translations/NeuraltoAmazon.txt
# ../Text/Translations/NeuralToAzure.txt
# ../Text/Translations/NeuraltoGoogle.txt
compare_files(master, ASR_Text)

```
